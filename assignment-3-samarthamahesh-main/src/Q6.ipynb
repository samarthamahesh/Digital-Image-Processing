```python
def find_root(conn, x):
    tmp = x
    while conn[tmp] != 0:
        tmp = conn[tmp]
        
    return tmp

def connected_components(im, neigh):
    im = np.int32(im)
    m, n = im.shape
    
    ret = np.zeros((m, n), dtype=int)
    p = 1
    conn = np.zeros((m*n), dtype=int)
    
    # Pass 1
    for i in range(m):
        for j in range(n):
            if im[i][j] == 1:
                arr = np.zeros(4, dtype=int)
                
                if j-1 >= 0:
                    arr[0] = ret[i][j-1]
                    
                if i-1 >= 0:
                    arr[1] = ret[i-1][j]
                    
                if neigh == 8:
                    if i-1 >= 0 and j-1 >= 0:
                        arr[2] = ret[i-1][j-1]
                        
                    if i-1 >= 0 and j+1 < n:
                        arr[3] = ret[i-1][j+1]
                
                arr = arr[arr != 0]
                
                if arr.shape[0] == 0:
                    ret[i][j] = p
                    p += 1
                else:
                    ret[i][j] = np.min(arr)
                    min_root = find_root(conn, ret[i][j])
                    
                    rep = arr[arr != ret[i][j]]
                    rep = np.unique(rep)
                    
                    if rep.shape[0] == 0:
                        continue
                        
                    for k in rep:
                        k_root = find_root(conn, k)
                        
                        if k_root != min_root:
                            conn[k_root] = ret[i][j]
                        
    # Pass 2
    for i in range(m):
        for j in range(n):
            if im[i][j] == 1:
                ret[i][j] = find_root(conn, ret[i][j])
                
    labels = np.unique(ret)
    for i in range(len(labels)):
        ret[ret == labels[i]] = i
    
    return labels.shape[0], ret

def erosion(im, se):
    im = np.int32(im)
    m, n = im.shape
    k, l = se.shape
    
    im = np.vstack((np.zeros(((k-1)//2, n+l-1)), np.hstack((np.zeros((m, (l-1)//2)), im, np.zeros((m, (l-1)//2)))), np.zeros(((k-1)//2, n+l-1))))
    ret = np.zeros((m, n))
    
    for i in range(m):
        for j in range(n):
            ret[i][j] = np.min(im[i:i+k, j:j+l][se == 1])
            
    return ret

def dilation(im, se):
    im = np.int32(im)
    m, n = im.shape
    k, l = se.shape
    
    im = np.vstack((np.zeros(((k-1)//2, n+l-1)), np.hstack((np.zeros((m, (l-1)//2)), im, np.zeros((m, (l-1)//2)))), np.zeros(((k-1)//2, n+l-1))))
    ret = np.zeros((m, n))
    
    for i in range(m):
        for j in range(n):
            ret[i][j] = np.max(im[i:i+k, j:j+l][se == 1])
            
    return ret

def circle_kernel(x):
    k, l = np.meshgrid(np.arange(x), np.arange(x))
    k = k - np.mean(k, axis=1)
    l = l - np.mean(l, axis=0)
    
    ret = np.sqrt(k**2 + l**2)
    
    return np.where(ret <= x//2, 1, 0)

def disc_kernel(x):
    out_k = circle_kernel(x)
    in_k = np.vstack((np.zeros((1, x)), np.hstack((np.zeros((x-2, 1)), circle_kernel(x-2), np.zeros((x-2, 1)))), np.zeros((1, x))))
    
    return out_k - in_k
```

```python
im = cv2.imread('../images/kidney.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(im, 130, 1, cv2.THRESH_BINARY)

kernel = disc_kernel(11)
opening = dilation(erosion(thresh, kernel), kernel)

kernel = disc_kernel(5)
closing = erosion(dilation(opening, kernel), kernel)

num_labels, labels_im = connected_components(closing, 8)

rgb = label2rgb(labels_im, bg_label=0)

plt.imshow(rgb)
plt.show()
```

#### 1. Given the label map of a scene, how do you upscale the label map by factor k?

For upscaling the label map by factor **k**, we generate the following matrix,

**[[k1, 0]\
[0, k2]]**

where, k1 and k2 scales the label map in horizontal and vertical directions.

As we are upscaling the label map, **k1 > 1** and **k2 > 1**.

As the number of output pixels are more than the original label map, we can use following methods to counter this,

1. **Interpolation**

    * In this method, we set the surrounding pixels (blank/unmapped pixels) to the current pixel value (filled/mapped pixel). This will reduce the blank spaces and makes the output more uniform (without black bars of unmapped pixels).
    
    
2. **Backward mapping**

    * In this method, we map back all the pixels in the scaled label map to original label map. So, each pixel in scaled label map will get mapped to a specific pixel in original label map. Hence, this will not result in any blank/unmapped pixels in output scaled label map. For this method, we use inverse of the scaling matrix to do backward mapping.
    
    Hence, the matrix will be,\
    **[[1/k1, 0]\
    [0, 1/k2]]**

```python
def label_scale(im, k1, k2):
    im = np.int32(im)
    m, n = im.shape
    
    # Coordinates in input image
    x, y = np.meshgrid(np.arange(n), np.arange(m))
    x = x - np.mean(x, axis=1, keepdims=True)
    y = y - np.mean(y, axis=0, keepdims=True)
    y = np.flip(y, axis=0)
    
    # Scale matrix
    mat = np.array([[k1, 0],
                   [0, k2]])
    
    # Inverse of scale matrix
    mat_inv = np.linalg.inv(mat)
    
    # Finding output image's size
    m1, n1 = k2*m, k1*n
        
    # Coordinates in output image
    x1, y1 = np.meshgrid(np.arange(n1), np.arange(m1))
    x1 = x1 - np.mean(x1, axis=1, keepdims=True)
    y1 = y1 - np.mean(y1, axis=0, keepdims=True)
    y1 = np.flip(y1, axis=0)
        
    ret = np.zeros((m1, n1))
    cm = m//2
    cn = n//2
    
    # Implementation of backward mapping
    for i in range(m1):
        for j in range(n1):
            tmp = np.array([[x1[i][j]], [y1[i][j]]])
            new = np.matmul(mat_inv, tmp)
            new = np.int32(new)
            xn, yn = cm-new[1][0], cn+new[0][0]
            
            if xn >= 0 and xn < m and yn >= 0 and yn < n:
                ret[i][j] = im[xn][yn]
            
    return np.uint8(ret)
```

```python
scale_map = label_scale(labels_im, 3, 3)
rgb = label2rgb(scale_map, bg_label=0)

plt.imshow(rgb)
plt.title("k1: 3, k2: 3")
plt.show()

scale_map = label_scale(labels_im, 3, 5)
rgb = label2rgb(scale_map, bg_label=0)

plt.imshow(rgb)
plt.title("k1: 3, k2: 5")
plt.show()

scale_map = label_scale(labels_im, 5, 3)
rgb = label2rgb(scale_map, bg_label=0)

plt.imshow(rgb)
plt.title("k1: 5, k2: 3")
plt.show()
```

#### 2. Given the label map of a scene, how do you rotate the label map by angle theta?

For rotating the label map by angle **theta**, we generate the following matrix,

**[[cos(theta), sin(theta)]\
[-sin(theta), cos(theta)]**

where theta is in **radians**. As I'm assuming _positive angle as clock-wise rotation_ I have the above matrix. If you assume the oppsite, i.e., _positive angle means anticlock-wise rotation_, then the matrix will look like,

**[[cos(theta), -sin(theta)]\
[sin(theta), cos(theta)]**

Again, as discussed before, many pixels in output label map will not be mapped and hence will remain black. This will result in unwanted black bars (due to unmapped pixels). This is caused due to discretised domain of images or label maps. To counter this issue, we can use backward mapping as stated before which would map each pixel in the required output label map to priginal label map, which means we have a valid mapped pixel for each pixel in output label map. Hence, we countered the problem with this method.

To account for increased size of rotated label map, we can find the output coordinates of the four corners of the original label map after being rotated by an angle theta. Then we can calculate the horizontal and vertical distances between the four corner pixels we have. Hence this will be the output label map's size. This is only if we want to retain the complete label map after rotating. If we want original size of label map, we will have to sacrifice some amount of pixels which will go out of the boundary (original label map's size).

```python
def label_rotate(im, deg, retain):
    im = np.int32(im)
    m, n = im.shape
    
    # Coordinates in input image
    x, y = np.meshgrid(np.arange(n), np.arange(m))
    x = x - np.mean(x, axis=1, keepdims=True)
    y = y - np.mean(y, axis=0, keepdims=True)
    y = np.flip(y, axis=0)
    
    # Rotation matrix
    mat = np.array([[np.cos(deg*np.pi/180), np.sin(deg*np.pi/180)],
                   [-np.sin(deg*np.pi/180), np.cos(deg*np.pi/180)]])
    
    # Inverse of rotation matrix
    mat_inv = np.linalg.inv(mat)
    
    # Finding output image's size
    if retain:
        m1, n1 = m, n
        
    else:
        tmp = np.array([[x[0][0], x[0][-1], x[-1][0], x[-1][-1]], [y[0][0], y[0][-1], y[-1][0], y[-1][-1]]])
        new = np.matmul(mat, tmp)
        
        n_min = np.min(new, axis=1)
        n_max = np.max(new, axis=1)
        
        n1, m1 = np.int32(n_max - n_min + 1)
        
    # Coordinates in output image
    x1, y1 = np.meshgrid(np.arange(n1), np.arange(m1))
    x1 = x1 - np.mean(x1, axis=1, keepdims=True)
    y1 = y1 - np.mean(y1, axis=0, keepdims=True)
    y1 = np.flip(y1, axis=0)
        
    ret = np.zeros((m1, n1))
    cm = m//2
    cn = n//2
    
    # Implementation of backward mapping
    for i in range(m1):
        for j in range(n1):
            tmp = np.array([[x1[i][j]], [y1[i][j]]])
            new = np.matmul(mat_inv, tmp)
            new = np.int32(new)
            xn, yn = cm-new[1][0], cn+new[0][0]
            
            if xn >= 0 and xn < m and yn >= 0 and yn < n:
                ret[i][j] = im[xn][yn]
            
    return np.uint8(ret)
```

```python
rot_map = label_rotate(labels_im, 45, True)
rgb = label2rgb(rot_map, bg_label=0)

plt.imshow(rgb)
plt.title('Angle: 45 deg')
plt.show()

rot_map = label_rotate(labels_im, -45, True)
rgb = label2rgb(rot_map, bg_label=0)

plt.imshow(rgb)
plt.title('Angle: -45 deg')
plt.show()
```
