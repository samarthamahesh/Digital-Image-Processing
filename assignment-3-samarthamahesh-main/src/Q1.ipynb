```python
s1 = np.array([[0, 0, 0],
              [np.nan, 1, np.nan],
              [1, 1, 1]])

s2 = np.array([[np.nan, 0, 0],
              [1, 1, 0],
              [np.nan, 1, np.nan]])
```

#### 1. Write code for the above skeletonization process and display the computed skeleton for the image thumb print.jpg. Note that the image will need to be binarized first ! You can use cv2.threshold or any other inbuilt function for binarization.

```python
def morphological_op(im, st_ele):
    m, n = im.shape
    k, l = st_ele.shape
    ret = np.copy(im)
    
    im = np.vstack((np.zeros(((k-1)//2, n+l-1)), np.hstack((np.zeros((m, (l-1)//2)), im, np.zeros((m, (l-1)//2)))), np.zeros(((k-1)//2, n+l-1))))
    
    for i in range(m):
        for j in range(n):
            if (k*l)-np.sum(np.isnan(st_ele)) == np.sum(st_ele == im[i:i+k, j:j+l]):
                ret[i][j] = 0
    
    return ret

def skeletonize(im, s1, s2):
    new = np.copy(im)
    
    while True:
        temp = morphological_op(new, s1)
        if np.all(new == temp):
            new = temp
            break
        new = temp
        
        temp = morphological_op(new, s2)
        if np.all(new == temp):
            new = temp
            break
        new = temp
        
        s1 = np.rot90(s1)
        s2 = np.rot90(s2)
        
    return new
```

```python
im = cv2.imread('../images/thumb_print.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(im, 127, 1, cv2.THRESH_BINARY_INV)

ret = skeletonize(thresh, s1, s2)
plt.imshow(ret, cmap='gray')
```

#### 2. Explain the role of structuring elements S1 , S2.

Structuring element S1 finds protrusions on the foreground. Hence, according to the algorithm proposed, S1 removes the protrusions on the foreground. Because skeleton of an image will have thin lines without any protrusions.

Structuring element S2 finds the square corners in the foreground. Hence, according to the algorithm proposed, S2 removes square corners from the foreground to thin the foreground. As foreground can have thick lines or shapes, they will have corners. Hence applying this structuring element, the square corners are removed.

#### 3. By analyzing the outputs at the end of each skeletonization pass, explain why multiple passes are/may be required.

```python
im = cv2.imread('../images/thumb_print.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(im, 127, 1, cv2.THRESH_BINARY_INV)

ret = skeletonize(thresh, s1, s2)
plt.imshow(ret, cmap='gray')
```

From above images for each iteration, we see that the proposed algorithm doesn't compute the skeleton of the image instantaneously. It requires several iterations to find the skeleton. One iteration of algorithm removes the protrusions and square corners which are present on the boundary of the foreground. This produces a thinner foreground because some pixels corresponding to protrusions and square corners. But this is not the skeleton, infact this has more protrusions and corners which are produced by removing some pixels on boundary in previous iteration. Hence, this is a iterative process and requires more iterations to get the thin skeleton of foreground image.
