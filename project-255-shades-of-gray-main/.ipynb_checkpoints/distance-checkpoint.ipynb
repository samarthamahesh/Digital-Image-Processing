```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.morphology import skeletonize 
from skimage.morphology import remove_small_objects
from skimage.morphology import binary_erosion
from scipy.ndimage import distance_transform_edt as DT
import cv2 as cv
import time
import plotly
```

```python
"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
FUNCTION    -    get_skel(img)

Inputs
img -> Image whose skeleton needs to be generated


Outputs
i3-> Skeleton of the image after removing extra branches

In this function, we obtain the skeleton of the image using lee method which helps in removal of additonal 
small branches.
"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
# Skeleton transform
def get_skel(img):
    
    # Thresholding image 
    _,i1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
    # Skeletonizing with removal of branches using "Lee Skeletonization"
    i2 = skeletonize(i1, method='lee')
    i3 = remove_small_objects(i2, min_size=9, connectivity=2)
    
    return i3

"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
FUNCTION    -    get_peri(img)

Inputs
img -> Image for which outline/contour has to be generated


Outputs
per -> Outline/Contour of the input image

In this function, we obtain the outline of the image by eroding the image and subtracting the eroded image from
the original image
"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
# Perimeter transform
def get_peri(img):
    
    # Thresholding image  
    _,thresh = cv.threshold(img,127,255,cv.THRESH_BINARY)
    # Boundary pixels by subtracting eroded image from original image
    ero=255*binary_erosion(thresh).astype(int)    
    per = thresh-ero
    
    return per
```

FUNCTION    -    get_dist(img)

Inputs\
img -> Image for which normalised distance transform image has to be generated


Outputs\
per -> Normal distance transform image\
We take the input image, and then compute the normalised distance using the following formula :



```python

# Custom distance transform
def get_dist(img):
    
    # Distance transform on Image boundary and Image skeleton 
    skl = get_skel(img)
    skel_dt = DT(skl==0)
    per = get_peri(img)
    per_dt, inds = DT(per==0, return_indices=True)
    
    # Dist from the skeleton from the closest point on contour
    dm3 = skel_dt[inds[0,:,:],inds[1,:,:]]

    # Modified text radius calculation 
    py = np.sort(skel_dt[per>0])
    px = np.asarray(range(round(np.size(py)*0.2),round(np.size(py)*0.8)+1))
    p=np.polyfit(px,py[px],1)
    c=np.size(py)*0.2*p[0]+p[1]
    dm3[dm3 < c] = c

    # Mean text radius
    mean_dist = (np.size(py)*0.5*p[0]+p[1])

    # normalised distance - final output
    _,thresh = cv.threshold(img,127,255,cv.THRESH_BINARY)
    modified_dist = np.where(thresh, 1-np.divide(per_dt,dm3), 1+np.divide(per_dt,mean_dist))
    
    return modified_dist
    
```

```python
img = cv.imread('shu-text.png',0)
img = cv.imread('new-text.png',0)
#_,thresh = cv.threshold(img,127,255,cv.THRESH_BINARY)

t = time.time()
output = get_dist(img)


plt.imshow(output,cmap='jet')
plt.colorbar()
plt.show()

print("Time taken for execution -",time.time()-t)
```
